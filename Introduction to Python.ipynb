# Threading and Scripting in Python

```python
from threading import *
class Demo:
    def show(self):
        for i in range(5):
            print("This is the child thread")
obj = Demo()
t = Thread(target = obj.show())
t.start()
for i in range(5):
    print("This is the parent thread")
```

# Multithreading 

```python
from threading import *
import time
class Demo:
    def num(self):
        for i in range(1,6):
            print("The number is", i)
            time.sleep(i)
        
    def double(self):
        for i in range(1,6):
            print("The double of the number is", 2*i)
            time.sleep(1)
            
    def square(self):
        for i in range(1,6):
            print("The Square of the number is", i*1)
            time.sleep(1)
            
obj = Demo()
t1 = Thread(target = obj.num)
t2 = Thread(target = obj.double)
t3 = Thread(target = obj.square)

t1.start()
time.sleep(0.2)
t2.start()
time.sleep(0.2)
t3.start()

t1.join()
t2.join()
t3.join()

print("This is the main thread")
```

```python
wine = pd.read_csv('wineQT.csv')
print(wine.columns)
```

```python
# Preprocessing Data
bins=(2, 6.5, 8)
group_names=['bad', 'good']
wine['quality']=pd.cut(wine['quality'], bins=bins, labels=group_names)
wine['quality'].unique()
```

```python
label_quality=LabelEncoder()
```

```python
wine['quality']=label_quality.fit_transform(wine['quality'])
```

```python
wine.head(10)
```

```python
wine['quality'].value_counts()
```

```python
print(wine['quality'].dtype)
```

```python
wine['quality'] = wine['quality'].astype('category')
```

```python
# Ensure the 'quality' column is categorical
wine['quality'] = wine['quality'].astype('category')

# Plot the count plot
sns.countplot(x='quality', data=wine)
plt.title('Quality Count')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
# Now seperate the dataset as response variable and feature variables
X=wine.drop('quality',axis=1)
Y=wine['quality']
```

```python
# Train and test splitting of data
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Applying Standard scaling to get optimized result

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

# Random Forest Classifier

```python
rfc=RandomForestClassifier(n_estimators=200)
rfc.fit(X_train, Y_train)
pred_rfc=rfc.predict(X_test)
```

```python
# Let's see how our model performed
print(classification_report(Y_test, pred_rfc))
print(confusion_matrix(Y_test, pred_rfc))
```

# SVM Classifier

```python
# Let's see how our model performed
print(classification_report(Y_test, pred_clf))
print(confusion_matrix(Y_test, pred_clf))
```

```python
mlpc=MLPClassifier(hidden_layer_sizes=(11,11,11),max_iter=500)
mlpc.fit(X_train, Y_train)
pred_mlpc=mlpc.predict(X_test)
```
